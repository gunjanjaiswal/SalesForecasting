Feature Engineering

```python

train['Sales_Rolling3'] = (
    train.groupby(['Store', 'Dept'])['Weekly_Sales']
    .shift(1)  
    .rolling(window=3)
    .mean()
)
```

```python
train['Sales_Lag1'] = (
    train.groupby(['Store', 'Dept'])['Weekly_Sales']
    .shift(1)  # Just the previous week's sales
)
```

```python
def extractDateFeatures(df):
    if not pd.api.types.is_datetime64_any_dtype(df['Date']):
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Weekday'] = df['Date'].dt.weekday
    df['Day'] = df['Date'].dt.day
    df['Quarter'] = df['Date'].dt.quarter
    df['Week'] = df['Date'].dt.isocalendar().week  #
    return df
```

```python
train = extractDateFeatures(train)
```

```python
train['Sales_Lag1'] = train['Sales_Lag1'].fillna(0)
train['Sales_Rolling3'] = train['Sales_Rolling3'].fillna(0)
```

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = ['Store', 'Dept', 'Type']  
label_encoders = {}

for col in cat_cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    label_encoders[col] = le
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import mutual_info_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

target_column = 'Weekly_Sales' 

X = train.drop(columns=[target_column])
y = train[target_column]

datetime_cols = X.select_dtypes(include=['datetime64', 'datetime']).columns
X = X.drop(columns=datetime_cols)

numeric_features = X.select_dtypes(include='number')
numeric_df = pd.concat([numeric_features, y], axis=1)

plt.figure(figsize=(12, 10))
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Numeric Features Only)')
plt.show()

target_corr = corr_matrix[target_column].drop(target_column)
low_corr_features = target_corr[abs(target_corr) < 0.1].index.tolist()
print("\n Features with low linear correlation to target:", low_corr_features)
```

```python
dates = train['Date']

X = train.drop(columns=[
    'Weekly_Sales', 'Date', 'Temperature', 'Fuel_Price', 'MarkDown1', 'MarkDown2',
    'MarkDown5', 'MarkDown4', 'Year', 'Month', 'Weekday', 'Day', 'Quarter'
])
y = train['Weekly_Sales']

# Split Date along with features and labels
from sklearn.model_selection import train_test_split

X_train, X_temp, y_train, y_temp, dates_train, dates_temp = train_test_split(
    X, y, dates, test_size=0.2, random_state=42
)

X_valid, X_test, y_valid, y_test, dates_valid, dates_test = train_test_split(
    X_temp, y_temp, dates_temp, test_size=0.5, random_state=42
)

dates_train = pd.to_datetime(dates_train).reset_index(drop=True)
dates_valid = pd.to_datetime(dates_valid).reset_index(drop=True)
dates_test = pd.to_datetime(dates_test).reset_index(drop=True)

print(f"Training set size: {X_train.shape}")
print(f"Validation set size: {X_valid.shape}")
print(f"Test set size: {X_test.shape}")

```

```python
# 3. Standard Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
```

```python
!pip install tensorflow scikit-learn

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_valid_scaled = scaler.transform(X_valid)
X_test_scaled = scaler.transform(X_test)

model = Sequential([
    Dense(1024, activation='relu', kernel_regularizer=regularizers.l2(0.001), input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.3),
    Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.3),
    Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dropout(0.2),
    Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
    Dense(1)
])


from tensorflow.keras.losses import Huber
model.compile(optimizer='adam', loss=Huber(), metrics=['mae'])

# Learning Rate Scheduler to reduce LR on plateau
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6)
# Early stopping to prevent overfitting and restore best weights
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(
    X_train_scaled, y_train,
    validation_data=(X_valid_scaled, y_valid),
    epochs=50,
    batch_size=32,
    callbacks=[lr_scheduler, early_stopping]
)

y_pred_nn = model.predict(X_valid_scaled)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_valid, y_pred_nn)
mae = mean_absolute_error(y_valid, y_pred_nn)
r2 = r2_score(y_valid, y_pred_nn)

print(f"Mean Squared Error: ",mse)
print(f"Mean Absolute Error:",mae)
print(f"R2 Score:", r2)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training vs Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Predict
y_pred_nn = model.predict(X_valid).flatten()

import matplotlib.pyplot as plt
import pandas as pd

y_valid = pd.Series(y_valid).reset_index(drop=True)
y_pred_nn = pd.Series(y_pred_nn).reset_index(drop=True)

# Indices
indices = range(len(y_valid))

plt.figure(figsize=(14,7))
plt.plot(indices, y_valid, label='Actual Sales', color='blue', linewidth=2)
plt.plot(indices, y_pred_nn, label='Predicted Sales (NN)', color='orange', linestyle='--', linewidth=2)

plt.fill_between(indices, y_valid, y_pred_nn, color='grey', alpha=0.2)

plt.title('Neural Network - Actual vs Predicted Weekly Sales', fontsize=16)
plt.xlabel('Sample Index')
plt.ylabel('Weekly Sales')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
